```
import numpy as np
import pandas as pd
from PIL import Image
from wordcloud import WordCloud, STOPWORDS, ImageColorGenerator
import matplotlib.pyplot as plt
%matplotlib inline
```

```
Wordclouddata = pd.read_csv("http://raw.githubusercontent.com/hadley/data-baby-names/master/baby-names.csv")
```

```
from wordcloud import WordCloud, STOPWORDS
wordcloud = WordCloud(
                          background_color='white',
                          stopwords=STOPWORDS,
                          max_words=100,
                          max_font_size=80,
                          random_state=42
                          ).generate(str(Wordclouddata2))

plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.margins(x=0, y=0)
plt.show()
```

```
Wordclouddata3 = Wordclouddata2.rename(columns={'name': ''})
```

```
from wordcloud import WordCloud, STOPWORDS
wordcloud = WordCloud(
                          background_color='white',
                          stopwords=STOPWORDS,
                          max_words=100,
                          max_font_size=80,
                          random_state=42
                          ).generate(str(Wordclouddata3))

plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.margins(x=0, y=0)
plt.show()
```
